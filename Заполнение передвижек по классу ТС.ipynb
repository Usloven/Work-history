```python
def chg (x, change = change):
    for i in change:
        if change[i] == x:
            x = i
    return x
    
        
```

```python
def fill_class_avto_regions(file, df, catts):
    
    df1 = df
    df1["Регион"] = df1["Регион"].apply(chg)
    df1["ФО"] = df1["ФО"].str.replace("ФО", "федеральный округ")
    df1["ФО"] = df1["ФО"].apply(chg)
    
    df1fo = df1[df1["Категория ТС"]==catts]
    df1fo = df1fo.groupby(["ФО"]).sum(numeric_only=True)
    
    df1reg = df1[df1["Категория ТС"]==catts]
    df1reg = df1reg.groupby(['Регион']).sum(numeric_only=True)
    
    df1rf = df1fo.reset_index()
    df1rf["Российская Федерация"] = "Российская Федерация"
    df1rf= df1rf.drop(columns = "ФО")
    df1rf = df1rf.groupby(["Российская Федерация"]).sum(numeric_only=True)
    
    ws = catts
    file2 = file
    report =  pd.read_excel(file2, sheet_name=ws, index_col=0, skiprows=1)
    report = check_col_name(report)
    
    for fo in df1fo.index:
        for col in df1fo:
            report.loc[fo, col] = df1fo.loc[fo, col]
    
    for reg in df1reg.index:
        for col in df1reg:
            report.loc[reg, col] = df1reg.loc[reg, col]
    
    for country in df1rf.index:
        for col in df1rf:
            report.loc[country, col] = df1rf.loc[country, col]  
    update_spreadsheet(file, report,  2, 3, catts)
```

```python
def check_col_name (report):
    for name  in report.columns:
        if name == "СН4":
            report.rename(columns={name: "CH4"}, inplace =True)
    return report
        
import re 

def has_cyrillic(text):
    return bool(re.search('[а-яА-Я]', text))
    
```

```python

def update_spreadsheet(path : str, _df, starcol : int = 1, startrow : int = 1, sheet_name : str ="ToUpdate"):
    '''
    
    :param path: Путь до файла Excel
    :param _df: Датафрейм Pandas для записи
    :param starcol: Стартовая колонка в таблице листа Excel, куда буду писать данные
    :param startrow: Стартовая строка в таблице листа Excel, куда буду писать данные
    :param sheet_name: Имя листа в таблице Excel, куда буду писать данные
    :return: 
    '''
    wb = ox.load_workbook(path)
    for row in wb[sheet_name]['B4:D300']:
        for cell in row:
            cell.value = None
    for ir in range(0, len(_df)):
        for ic in range(0, len(_df.iloc[ir])):
            wb[sheet_name].cell(startrow + ir, starcol + ic).value = _df.iloc[ir][ic]
    wb.save(path)


```
