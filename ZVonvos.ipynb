```python
list_code = []
list_zv = []
list_file = []
def fill_zv (file1, list_code, list_zv, list_file):
    wb = ox.load_workbook(file1)
    for catts in wb.sheetnames:
        list_code.append(wb[catts]["B2"].value)
        list_zv.append(catts)
        list_file.append(file1)
```

```python
directory = r'C:\Users\aleksandrov.vd\Documents\Выгрузки\ОНВОС ЗВ по коду'
file_list = []
towns = []
city = []
# iterate over files in
# that directory
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    # checking if it is a file
    if os.path.isfile(f) and 'xlsx' in f:
        file_list.append(f)
        towns.append(filename)
```

```python
for i in towns:
    i = i.split(".")[0]
    city.append(i)
```

```python
for file in file_list:
    fill_zv (file, list_code, list_zv, list_file)
```

```python
list_zv_total = []
for i in range(len(list_code)):
    list_zv_total.append(list_code[i]+";"+list_zv[i]+";" +list_file[i])
    
```

```python
df_zv = pd.DataFrame({'zv':list_zv_total})
df_zv["code"] = df_zv["zv"].apply(lambda x: x.split(";")[0])
df_zv["name"] = df_zv["zv"].apply(lambda x: x.split(";")[1])
df_zv["file"] = df_zv["zv"].apply(lambda x: x.split(";")[2])
df_zv
```

```python
df_zv = df_zv.groupby(["code", "name"]).count().reset_index()
```
