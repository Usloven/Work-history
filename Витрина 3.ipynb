```python
Q1 = df["Объем продукции"].quantile(q=.25)
Q3 = df["Объем продукции"].quantile(q=.75)
```

```python
quartiles = df["Объем продукции"].quantile([0.25, 0.75])
iqr = quartiles[0.75] - quartiles[0.25]
print(iqr)
```

```python
data_clean = df[~((df["Объем продукции"] < (Q1-1.5*iqr)) | (df["Объем продукции"] > (Q3+1.5*iqr)))]
```

```python
data_clean = data_clean[data_clean["Объем продукции"]>0]
```
