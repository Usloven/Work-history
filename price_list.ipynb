```python
import numpy as np
import pandas as pd
```

```python
file2 = r"смешанный.xlsx"
df2 = pd.read_excel(file2)
```

```python
file = r"ОБЪЕДЕНИТЬ НАДО.xlsx"
df = pd.read_excel(file )
```

```python
df = df.rename(columns= {"Column12":"Column11"})
```

```python
df10 = df.copy()
```

```python
df10=df10[["Column5", "Column11"]]
```

```python
df10 = df10.set_index('Column5')
```

```python
inn = df10.to_dict()["Column11"]
```

```python
inn
```

```python
df_comb1 = df2.merge(df, left_on="ИНН1", right_on = "Column5", how = "right" )
df_comb1  
    
```

```python
def fill_inn(x):
    if x > 0 and x in inn.keys():
        return inn[x]
    else:
        return 0
```

```python
df_comb1 = df_comb1[["Column11","Column5", "Column2", "ИНН2", "Компания2" ]]
df_comb1 = df_comb1.fillna(0)
```

```python
df_comb1["Выручка_доч"] = df_comb1["ИНН2"].apply(fill_inn)
```

```python
df_comb1
```

```python
df_comb1 = df_comb1.rename(columns = {"Column2":"Название_род", "Column5":"ИНН_род", "Column11":"Выручка_род", "Компания2":"Назв1_доч"})
```

```python
df_comb1
```

```python
df_comb1 = df_comb1[["Название_род", "ИНН_род", "Выручка_род", "ИНН2", "Назв1_доч", "Выручка_доч" ]]
```

```python
df_comb2 = df2.merge(df, left_on="ИНН2", right_on = "Column5", how = "right" )
df_comb2
```

```python
df_comb2 = df_comb2[["Column11","Column5", "Column2", "ИНН1", "Компания1" ]]
df_comb2 = df_comb2.fillna(0)
```

```python
df_comb2["Выручка_доч"] = df_comb2["ИНН1"].apply(fill_inn)
```

```python
df_comb2
```

```python
df_comb2 = df_comb2.rename(columns = {"Column2":"Название_род", "Column5":"ИНН_род", "Column11":"Выручка_род", "Компания1":"Назв1_доч"})
```

```python
df_comb2 = df_comb2.rename(columns = {"ИНН1":"ИНН2"})
```

```python
df_comb2 = df_comb2[["Название_род", "ИНН_род", "Выручка_род", "ИНН2", "Назв1_доч", "Выручка_доч" ]]
```

```python
df_comb1
```

```python
df_comb = pd.concat([df_comb1, df_comb2], ignore_index=True)
```
