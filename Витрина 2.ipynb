```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = "data-1686313174126.csv"
df = pd.read_csv(file, index_col=0)
df
```

```python
dftest  = df[["annual_value", "name", "region"]]
dftest = dftest.groupby(["region", "name"]).sum()
dfmean = dftest.reset_index()\
                .groupby(["name"]).mean()
dfmean = dfmean.sort_values(by='annual_value', ascending=False)
dfmean = dfmean.head(50)
```

```python
dfmean
```

```python
dftest
```

```python

dftest = dftest.join(dfmean, lsuffix='_caller', rsuffix='_other', how='right')
dftest
```

```python
dftest = dftest[dftest["annual_value_caller"]>dftest["annual_value_other"]]
```

```python
dftest
```

```python
dftest = dftest.reset_index().pivot(columns='region',index='name',values='annual_value_caller')
```

```python
dftest = dftest.fillna(0)
```

```python
plt.figure(figsize=(100,100))

# Add title
plt.title("Hotmap polutant")

# Heatmap showing average arrival delay for each airline by month
sns.heatmap(data=dftest, annot=False, vmin=50, vmax=100)

# Add label for horizontal axis
plt.xlabel("Airline")
```
